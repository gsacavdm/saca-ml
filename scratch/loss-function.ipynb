```python
import sklearn
from sklearn import datasets
from sklearn import preprocessing
```

```python
n = 500
train_x = X2[:n]
train_y = y2[:n]

test_x = X2[n:]
test_y = y2[n:]
```

```python
from tensorflow import keras
```

```python
model = keras.models.Sequential([
    keras.layers.Dense(25, input_dim=20, activation=keras.activations.relu, kernel_initializer='he_uniform'),
    keras.layers.Dense(1, activation=keras.activations.linear)
])
```

```python
model.compile(optimizer=keras.optimizers.SGD(lr=0.01, momentum=0.9), loss=keras.losses.mean_squared_error)
```

```python
history = model.fit(train_x, train_y, validation_data=(test_x, test_y), epochs=100)
```

```python
train_mse = model.evaluate(train_x, train_y)
```

```python
test_mse = model.evaluate(test_x, test_y)
```

```python
print("Train MSE: %.5f Test MSE: %.5f" % (train_mse, test_mse))
```

```python
plt.title("Loss / Mean Squared Error")
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```
